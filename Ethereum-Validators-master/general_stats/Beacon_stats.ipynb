## participation rate

```python
import pandas as pd
import os
import time

from tqdm import tqdm

import argparse

from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC

wd = webdriver.Chrome('chromedriver',options=chrome_options)
	# Call url through selenium driver
wd.get("https://beaconcha.in/charts/participation_rate")
    
# Parse dates and values in those dates

dates = wd.execute_script("return Highcharts.charts[0].series[0].xData");
part_rate = wd.execute_script("return Highcharts.charts[0].series[0].yData");

df = pd.DataFrame({'Dates': dates, 'part_rate': part_rate })

		# Store to dataframe and save it in current folder
df.to_csv('./output.csv', index=False)

		# Don't forget ot quit!
wd.quit()
```

```python
from datetime import datetime

df['date'] = pd.to_datetime(df['Dates'],unit='ms')
```

```python
import plotly.express as px

fig = px.line(df, x="date", y="part_rate", title='Participation Rate - until "2022-12-05 03:28:23')
df = df.sort_values(by="date")
fig.show()
```

##validators

```python
wd = webdriver.Chrome('chromedriver',options=chrome_options)
	# Call url through selenium driver
wd.get("https://beaconcha.in/charts/validators")
    
# Parse dates and values in those dates
dates = wd.execute_script("return Highcharts.charts[0].series[0].xData");
validators= wd.execute_script("return Highcharts.charts[0].series[0].yData");

validator = pd.DataFrame({'Dates': dates, 'validators': validators })
validator['date'] = pd.to_datetime(validator['Dates'],unit='ms')

wd.quit()
```

```python
fig2 = px.line(validator, x="date", y="validators", title='Validators (number) until 2022-12-11')
validator = validator.sort_values(by="date")
fig2.show()
```

##average_balance

```python
wd = webdriver.Chrome('chromedriver',options=chrome_options)
	# Call url through selenium driver
wd.get("https://beaconcha.in/charts/average_balance")
    
# Parse dates and values in those dates
dates = wd.execute_script("return Highcharts.charts[0].series[0].xData");
balance= wd.execute_script("return Highcharts.charts[0].series[0].yData");

balance = pd.DataFrame({'Dates': dates, 'balance': balance})
balance['date'] = pd.to_datetime(balance['Dates'],unit='ms')

wd.quit()
```

```python
import plotly.express as px
fig = px.histogram(balance, x="balance", title='Validators - avg. Balance until 2022-12-11'  )
fig.show()
```
